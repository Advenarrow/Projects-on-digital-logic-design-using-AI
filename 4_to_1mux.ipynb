```python
%%writefile mux4to1.v

// Define a module named mux4to1
module mux4to1 (
  input in0,   // First data input
  input in1,   // Second data input
  input in2,   // Third data input
  input in3,   // Fourth data input
  input [1:0] sel, // 2-bit select line
  output out   // 1-bit output
);

  // Dataflow modeling using a continuous assignment (assign statement)
  // The output 'out' is assigned based on the value of 'sel'
  // This uses a nested conditional (ternary) operator
  assign out = (sel == 2'b00) ? in0 :  // If sel is 00, output in0
               (sel == 2'b01) ? in1 :  // Else if sel is 01, output in1
               (sel == 2'b10) ? in2 :  // Else if sel is 10, output in2
               in3;                   // Else (if sel is 11), output in3

endmodule // End of mux4to1 module
```

This code block generates the Verilog code for a 4-to-1 multiplexer (4:1 MUX).

The 4:1 MUX is a fundamental combinational circuit with the following characteristics:

*   **Four 1-bit data inputs:** `in0`, `in1`, `in2`, `in3`
*   **Two 1-bit select lines:** `sel[1:0]` (allowing for 4 possible selections: 00, 01, 10, 11)
*   **One 1-bit output:** `out`

The logic is as follows:
*   `out = in0` when `sel = 2'b00`
*   `out = in1` when `sel = 2'b01`
*   `out = in2` when `sel = 2'b10`
*   `out = in3` when `sel = 2'b11`

The Verilog code is designed to be clean, synthesizable, use dataflow modeling (`assign` statement), and includes inline comments for clarity.

```python
%%writefile mux4to1_tb.v

`timescale 1ns / 1ps // Define timescale for simulation

module mux4to1_tb;

  // Declare signals for inputs and output
  reg in0, in1, in2, in3; // 1-bit data inputs
  reg [1:0] sel;        // 2-bit select line
  wire out;             // 1-bit output

  // Instantiate the Device Under Test (DUT)
  mux4to1 dut (
    .in0(in0),
    .in1(in1),
    .in2(in2),
    .in3(in3),
    .sel(sel),
    .out(out)
  );

  // Dump waves for verification (optional, requires a waveform viewer like GTKWave)
  initial begin
    $dumpfile("mux4to1_tb.vcd"); // Specify the name of the VCD file
    $dumpvars(0, mux4to1_tb);    // Dump all variables in the testbench
  end

  // Test stimulus generation
  initial begin
    // Initialize inputs to known values
    in0 = 1'b0;
    in1 = 1'b0;
    in2 = 1'b0;
    in3 = 1'b0;
    sel = 2'b00;

    // Monitor changes in inputs and output
    $monitor("Time=%0t | in0=%b in1=%b in2=%b in3=%b | sel=%b | out=%b",
             $time, in0, in1, in2, in3, sel, out);

    #10; // Wait 10ns for initialization

    // Test Case 1: sel = 00, output should be in0
    in0 = 1'b1; in1 = 1'b0; in2 = 1'b0; in3 = 1'b0; sel = 2'b00; #10;
    in0 = 1'b0; in1 = 1'b1; in2 = 1'b1; in3 = 1'b1; sel = 2'b00; #10;

    // Test Case 2: sel = 01, output should be in1
    in0 = 1'b0; in1 = 1'b1; in2 = 1'b0; in3 = 1'b0; sel = 2'b01; #10;
    in0 = 1'b1; in1 = 1'b0; in2 = 1'b1; in3 = 1'b1; sel = 2'b01; #10;

    // Test Case 3: sel = 10, output should be in2
    in0 = 1'b0; in1 = 1'b0; in2 = 1'b1; in3 = 1'b0; sel = 2'b10; #10;
    in0 = 1'b1; in1 = 1'b1; in2 = 1'b0; in3 = 1'b1; sel = 2'b10; #10;

    // Test Case 4: sel = 11, output should be in3
    in0 = 1'b0; in1 = 1'b0; in2 = 1'b0; in3 = 1'b1; sel = 2'b11; #10;
    in0 = 1'b1; in1 = 1'b1; in2 = 1'b1; in3 = 1'b0; sel = 2'b11; #10;

    // Test all inputs high
    in0 = 1'b1; in1 = 1'b1; in2 = 1'b1; in3 = 1'b1; sel = 2'b00; #10;
    in0 = 1'b1; in1 = 1'b1; in2 = 1'b1; in3 = 1'b1; sel = 2'b01; #10;
    in0 = 1'b1; in1 = 1'b1; in2 = 1'b1; in3 = 1'b1; sel = 2'b10; #10;
    in0 = 1'b1; in1 = 1'b1; in2 = 1'b1; in3 = 1'b1; sel = 2'b11; #10;

    // Test all inputs low
    in0 = 1'b0; in1 = 1'b0; in2 = 1'b0; in3 = 1'b0; sel = 2'b00; #10;
    in0 = 1'b0; in1 = 1'b0; in2 = 1'b0; in3 = 1'b0; sel = 2'b01; #10;
    in0 = 1'b0; in1 = 1'b0; in2 = 1'b0; in3 = 1'b0; sel = 2'b10; #10;
    in0 = 1'b0; in1 = 1'b0; in2 = 1'b0; in3 = 1'b0; sel = 2'b11; #10;

    $finish; // End simulation
  end

endmodule
```

This section generates the Verilog testbench (`mux4to1_tb.v`) for the 4-to-1 multiplexer (`mux4to1.v`).

The testbench will perform the following actions:

*   **Instantiate the Device Under Test (DUT):** It includes an instance of our `mux4to1` module.
*   **Define Stimuli:** It declares registers for the inputs (`in0`, `in1`, `in2`, `in3`, `sel`) and a wire for the output (`out`).
*   **Apply Test Vectors:** An `initial` block will apply various combinations of input data and select line values to thoroughly test the MUX's functionality.
*   **Monitor Output:** `$monitor` will be used to display the input values (in0, in1, in2, in3, sel) and the corresponding output (`out`) for each test case.
*   **Time Control:** `#` delays are used to advance simulation time between applying new test vectors.
*   **Waveform Dumping (Optional but Recommended):** `$dumpfile` and `$dumpvars` are included to generate a VCD (Value Change Dump) file (`mux4to1_tb.vcd`). This file can be viewed with a waveform viewer (like GTKWave) to visually verify the circuit's behavior.
*   **Simulation Termination:** `$finish` will end the simulation after all test cases have been applied.

```python
%%bash

echo "Compiling Verilog module and testbench..."
# Compile the 4-to-1 MUX module and its testbench
# -o mux_sim.out specifies the output simulation executable name
iverilog -o mux_sim.out mux4to1.v mux4to1_tb.v

# Check if compilation was successful
if [ $? -eq 0 ]; then
  echo "Compilation successful. Running simulation..."
  # Run the compiled simulation using vvp (Verilog Virtual Machine)
  # The output is redirected to both stdout and a log file (sim.log)
  vvp mux_sim.out | tee sim.log

  echo "Simulation complete. Checking for waveform dump file..."
  # List the generated VCD file to confirm its creation and size
  ls -lh mux4to1_tb.vcd

  if [ $? -eq 0 ]; then
    echo "Waveform dump file (mux4to1_tb.vcd) successfully generated."
    echo "You can download 'mux4to1_tb.vcd' and view it with GTKWave for visual analysis."
  else
    echo "Error: Waveform dump file (mux4to1_tb.vcd) not found. Check testbench for $dumpfile and $dumpvars."
  fi
else
  echo "Compilation failed. Please check your Verilog code for errors."
fi
```

This section explains how to compile the Verilog module and testbench using `iverilog` and then simulate it using `vvp`.

1.  **Compile Verilog Code:** The `iverilog` command compiles the `mux4to1.v` module and `mux4to1_tb.v` testbench into an executable simulation file named `mux_sim.out`.
    *   `iverilog -o mux_sim.out mux4to1.v mux4to1_tb.v`
    *   `-o mux_sim.out` specifies the output file name for the compiled simulation.

2.  **Run Simulation:** The `vvp` command executes the compiled simulation file. The output is redirected to both the console and a log file (`sim.log`) using `tee sim.log`.
    *   `vvp mux_sim.out | tee sim.log`

3.  **Waveform Dump File:** The testbench is configured to generate a Value Change Dump (VCD) file named `mux4to1_tb.vcd`. This file contains the signal changes over time during the simulation.

4.  **Verification:** The script checks if the compilation was successful and then confirms the creation of the `mux4to1_tb.vcd` file. This VCD file can be viewed with a waveform viewer (like GTKWave) for visual verification.
